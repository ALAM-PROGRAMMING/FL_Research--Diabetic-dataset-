```python
us_columns = [
    "Diabetes_binary", "Age", "Sex", "BMI", "HighBP",
    "HighChol", "Smoker", "HvyAlcoholConsump",
    "PhysActivity", "AnyHealthcare"
]

india_columns = [
    "Diabetes_Status", "Age", "Gender", "BMI",
    "Hypertension", "Cholesterol_Level",
    "Smoking_Status", "Alcohol_Intake",
    "Physical_Activity", "Health_Insurance"
]

df_us = df_us[us_columns]
df_india = df_india[india_columns]

print("US Shape After Filtering:", df_us.shape)
print("India Shape After Filtering:", df_india.shape)
```

 4. Rename India Columns

```python
rename_mapping = {
    "Diabetes_Status": "Diabetes_binary",
    "Gender": "Sex",
    "Hypertension": "HighBP",
    "Cholesterol_Level": "HighChol",
    "Smoking_Status": "Smoker",
    "Alcohol_Intake": "HvyAlcoholConsump",
    "Physical_Activity": "PhysActivity",
    "Health_Insurance": "AnyHealthcare"
}

df_india.rename(columns=rename_mapping, inplace=True)

print("India columns renamed.")
print("\nIndia Columns After Rename:")
print(df_india.columns.tolist())
```

```python
print("US Age unique values:")
print(sorted(df_us["Age"].unique()))

print("\nIndia Age min & max:")
print("Min:", df_india["Age"].min())
print("Max:", df_india["Age"].max())
```

 5. Align Age Buckets

```python
def age_to_bucket(age):
    if age <= 24:
        return 1
    elif age <= 29:
        return 2
    elif age <= 34:
        return 3
    elif age <= 39:
        return 4
    elif age <= 44:
        return 5
    elif age <= 49:
        return 6
    elif age <= 54:
        return 7
    elif age <= 59:
        return 8
    elif age <= 64:
        return 9
    elif age <= 69:
        return 10
    elif age <= 74:
        return 11
    elif age <= 79:
        return 12
    else:
        return 13

# Apply only to India dataset
df_india["Age"] = df_india["Age"].apply(age_to_bucket)

print("Age alignment complete.")
print("India Age unique buckets:", sorted(df_india["Age"].unique()))
```

```python
print("US Missing Values:")
print(df_us.isna().sum())

print("\nIndia Missing Values:")
print(df_india.isna().sum())
```

```python
print("Unique values in India HvyAlcoholConsump:")
print(df_india["HvyAlcoholConsump"].unique())
```

Encode HvyAlcoholConsump (India)

```python
alcohol_mapping = {
    "Moderate": 0,
    "High": 1
}

df_india["HvyAlcoholConsump"] = df_india["HvyAlcoholConsump"].map(alcohol_mapping)

print("Alcohol column encoded.")
print(df_india["HvyAlcoholConsump"].unique())
```

 6. Median Imputation

```python
# 1. Dictionary to map complex strings to binary integers
mapping_rules = {
    "Diabetes_binary": {"No": 0, "Yes": 1},
    "Sex": {"Female": 0, "Male": 1, "Other": 0}, # US baseline defines 0=Female, 1=Male
    "HighBP": {"No": 0, "Yes": 1},
    "Smoker": {"Never": 0, "Former": 1, "Current": 1}, # US baseline defines 1 as "Ever smoked"
    "PhysActivity": {"Low": 0, "Medium": 1, "High": 1}, # US baseline defines 1 as "Any physical activity"
    "AnyHealthcare": {"No": 0, "Yes": 1}
}

# Apply mappings to the Indian dataset
df_india.replace(mapping_rules, inplace=True)
```

6.A. Imputing missing values (all columns are properly numeric now)

```python
df_us = df_us.fillna(df_us.median(numeric_only=True))
df_india = df_india.fillna(df_india.median(numeric_only=True))
```

6.B. Force both entire DataFrames to strict integers (Removing 1.0 / 0.0 floats)


```python
# 3. Force both entire DataFrames to strict integers (Removes 1.0 / 0.0 floats)
df_us = df_us.astype(int)
df_india = df_india.astype(int)

print("Data transformation complete. Missing values remaining:")
print(df_india.isna().sum().sum())
```

 7. Export Clean Datasets

```python
df_us.to_csv("../data/processed/us_cleaned_aligned.csv", index=False)
df_india.to_csv("../data/processed/india_cleaned_aligned.csv", index=False)

print("Cleaned datasets exported successfully yes.")
```

### 🏁 Data Engineering Summary: What We Just Did

Think of this notebook as a universal translator. We started with two totally different healthcare datasets—one from the US and one from India. They had different columns, different ways of counting age, and different ways of saying "Yes" or "No". 

**In simple terms, here is what we accomplished:**
1. **The Great Filter:** We kept only the 10 vital health clues (like BMI and Age) that both datasets shared, and threw away the rest.
2. **Speaking the Same Language:** We converted English words (like "Male/Female" or "Yes/No") into pure math (0s and 1s) because neural networks cannot read text.
3. **Filling the Potholes:** We found missing health records and safely filled them in using the "median" (the most normal, middle-ground number) so our models don't crash.
4. **The Final Polish:** We forced every single number to be a clean, solid integer (e.g., changing 1.0 to 1). 

**The Result:** Two perfectly matched datasets, speaking the exact same mathematical language, ready for machine learning!
