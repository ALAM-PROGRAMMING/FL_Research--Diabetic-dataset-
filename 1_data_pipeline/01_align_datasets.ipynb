## 1. Environment Setup

```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split

np.random.seed(42)
```

## 2. Load CDC BRFSS 2015 Dataset

```python
dataset_path = "../data/raw/diabetes_binary_health_indicators_BRFSS2015.csv"
df = pd.read_csv(dataset_path)

print(f"Loaded dataset shape: {df.shape}")
```

## 3. Strict Quality Assurance (QA) Check

```python
print("Missing Values by Column:")
print(df.isna().sum())

print("\nData Types by Column:")
print(df.dtypes)
```

## 4. Stratified 50/50 Split (New York vs. Texas)

```python
# Stratified split using the target variable 'Diabetes_binary' to ensure equal distribution
df_ny, df_tx = train_test_split(
    df, 
    test_size=0.5, 
    stratify=df["Diabetes_binary"], 
    random_state=42
)

print(f"Hospital NY Dataset Shape: {df_ny.shape}")
print(f"Hospital TX Dataset Shape: {df_tx.shape}")
```

## 5. Export Datasets to Processed Directory

```python
os.makedirs("../data/processed", exist_ok=True)

ny_out = "../data/processed/Hospital_NY.csv"
tx_out = "../data/processed/Hospital_TX.csv"

df_ny.to_csv(ny_out, index=False)
df_tx.to_csv(tx_out, index=False)

print(f"Exported NY data to: {ny_out}")
print(f"Exported TX data to: {tx_out}")
```
