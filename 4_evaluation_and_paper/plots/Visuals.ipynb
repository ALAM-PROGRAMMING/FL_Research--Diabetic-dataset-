```python
import matplotlib.pyplot as plt
import numpy as np

# IEEE Publication Style Formatting
plt.style.use('default')
fig, ax = plt.subplots(figsize=(10, 6))

# Data Arrays
models = ['Random Forest\n(Centralized)', 'Federated MLP\n(Our FL Network)', 'Logistic Regression\n(Centralized)', 'XGBoost\n(Centralized)', 'Voting Ensemble\n(Centralized)']
aucs = [0.792, 0.819, 0.820, 0.822, 0.824]

# Colors: Highlight the Federated model in a strong color, keep baselines neutral grey for academic printing
colors = ['#d3d3d3', '#2ca02c', '#d3d3d3', '#d3d3d3', '#d3d3d3']
edgecolors = ['#333333', '#000000', '#333333', '#333333', '#333333']

# Generate Bars
bars = ax.bar(models, aucs, color=colors, edgecolor=edgecolors, linewidth=1.5, width=0.6)

# Add exact scores on top of the bars
for bar in bars:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval + 0.002, f'{yval:.3f}', 
            ha='center', va='bottom', fontsize=12, fontweight='bold')

# Formatting the Chart
ax.set_ylim(0.75, 0.84) # Zooming the Y-axis to clearly show the marginal difference
ax.set_ylabel('ROC-AUC Score', fontsize=14, fontweight='bold')
ax.set_title('Diagnostic Power: Privacy-Preserving FL vs. Centralized Baselines', fontsize=16, fontweight='bold', pad=20)
ax.grid(axis='y', linestyle='--', alpha=0.7)
ax.tick_params(axis='x', labelsize=11)
ax.tick_params(axis='y', labelsize=12)

# Save the figure as a high-res PNG for your IEEE manuscript
plt.tight_layout()
plt.savefig('Federated_vs_Centralized_Results.png', dpi=300, bbox_inches='tight')
plt.show()
```
